```python
import requests
from datetime import datetime
```

```python
def log_error(function_name, error_type, message):
    """
    Writes timestamped error messages to error_log.txt
    File is opened in append mode so logs accumulate.
    """

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_message = (
        f"[{timestamp}] ERROR in {function_name}: "
        f"{error_type} — {message}\n"
    )

    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_message)
```

```python
# Task 1 — File Read & Write Basics
# Part A — Write to file

notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]

# Write Mode
with open("python_notes.txt", "w" , encoding="utf-8") as file:
    for line in notes:
        file.write(line + "\n")

print("File written successfully")

# Append Mode
with open("python_notes.txt", "a", encoding="utf-8" ) as file:
    file.write("Topic 6: Functions are reusable blocks of code used to perform specific tasks.\n")
    file.write("Topic 7: APIs allow program to communicate.\n")

print("Lines appended")



# Part B — Read file

print("\nReading file contents:\n")

try: 
    with open("python_notes.txt" , "r" , encoding="utf-8") as file:
      lines = file.readlines()
      
      count = 0

      for line in lines:
          count+=1
          print(f"{count}. {line.strip()}")

      print("\nTotal no of lines:" , count)
      
      keyword = input("\nEnter keyword to search")

      found =False
      for line in lines:
          if keyword.lower() in line.lower():
              print(f"Enter a keyword to search: {keyword}")
              print(line.strip())
              found = True

      if not found:
          print("No matching lines found")

except Exception as e:
    log_error("file_read", type(e).__name__, str(e))
```

```python
# Task 2 — API Integration
# Step 1 — Fetch and Display Products: 

print("\n---API Integration---")

BASE_URL = "https://dummyjson.com/products"

def fetch_products():
    """
    Fetch first 20 products and display in table format.
    """

    try:
        url = f"{BASE_URL}?limit=20"

        response = requests.get(url, timeout=5)

        data = response.json()

        products = data["products"]

        print(
            "ID   | Title     | Category        | Price     | Rating"
        )
        print("-" * 60)

        for product in products:
            print(
                product["id"],
                "|",
                product["title"],
                "|",
                product["category"],
                "| $",
                product["price"],
                "|",
                product["rating"]
            )

        return products
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
        log_error(
            "fetch_products",
            "ConnectionError",
            "Unable to connect"
        )
    
    except requests.exceptions.Timeout:
        print("Timeout occurred.")
        log_error(
            "search_laptops",
            "Timeout",
            "Slow response"
        )

    except Exception as e:
        print("Unexpected error:", e)
        log_error(
            "fetch_products",
            type(e).__name__,
            str(e)
        )


# Step 2 — Filter and Sort:
def filter_and_sort(products):

    """
    Filter rating >= 4.5 and sort by price descending.
    """

    print("\nFiltered and Sorted Products: \n")
    
    filtered=[]
    
    print(f"\nProducts whose rating>=4.5:\n")
    for p in products:
        if p["rating"] >=4.5:
            filtered.append(p)
            print(p["title"],
            "- $",
            p["price"],
            "- Rating:",
            p["rating"])
    

    print("\nFiltered list by price in descending order:\n")
    sorted_list = sorted(
        filtered,
        key=lambda x: x["price"],
        reverse=True
    )

    for item in sorted_list:
        print(
            item["title"],
            "- $",
            item["price"],
            "- Rating:",
            item["rating"]
        )

# Step 3 — Search by Category:
def search_laptops():
    """
    Fetch laptops category.
    """

    try:
        url = f"{BASE_URL}/category/laptops"

        response = requests.get(url, timeout=5)

        data = response.json()

        print("\nLaptop Products:\n")

        for item in data["products"]:
            print(
                item["title"],
                "- $",
                item["price"]
            )

    except requests.exceptions.ConnectionError:
        print("Connection failed.")
        log_error(
            "search_laptops",
            "ConnectionError",
            "Network issue"
        )

    except requests.exceptions.Timeout:
        print("Timeout occurred.")
        log_error(
            "search_laptops",
            "Timeout",
            "Slow response"
        )

    except Exception as e:
        print(e)
        log_error(
            "search_laptops",
            type(e).__name__,
            str(e)
        )

# Step 4 — POST Request (Simulated):

def create_product():
    """
    Send POST request to simulate product creation.
    """

    try:

        url= f"{BASE_URL}/add"

        payload = {
              "title": "My Custom Product",
              "price": 999,
              "category": "electronics",
              "description": "A product I created via API"
        }
         
        response = requests.post(
            url,
            json=payload,
            timeout=5
        )

        print("\nPOST Response:")
        print(response.json())

    except requests.exceptions.ConnectionError:
        print("Connection failed.")
        log_error(
            "create_product",
            "ConnectionError",
            "POST failed"
        )

    except requests.exceptions.Timeout:
        print("Timeout occurred.")
        log_error(
            "create_product",
            "Timeout",
            "POST timeout"
        )

    except Exception as e:
        print(e)
        log_error(
            "create_product",
            type(e).__name__,
            str(e)
        )
    

products_data = fetch_products()

if products_data:
    filter_and_sort(products_data)

search_laptops()
create_product()

```

```python
# Task 3 — Exception Handling

print("\n--- TASK 3 : EXCEPTION HANDLING---")

# Part A : Guarded Calculator

def safe_divide(a,b):

    try:
        return a/b
    
    except ZeroDivisionError:
        return "Error: Cannot Divide By Zero"
    
    except TypeError:
        return "Error: Invalid Input Types"
    

print(safe_divide(10,2))
print(safe_divide(10,0))
print(safe_divide("ten",2))


# Part B — Guarded File Reader:
 
def read_file_safe(filename):

    try:

        with open(filename, "r" , encoding="utf-8") as file:
            content = file.read()
            return content
        
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")

    finally:
        print("\nFile operation Attempt Complete")

print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))



# Part C & D — Robust API Calls:

while True:
    print(f"Enter Product ID :")
    user_input = input("\nEnter Product ID (1 -100)")
    print(f"Product ID entered: {user_input.lower(
        
    )}")
    if user_input.lower() == "quit":
        break

    if not user_input.isdigit():
        print("Please enter a valid Number")
        continue

    product_id = int(user_input)

    if product_id < 1 or product_id >100:
        print("ID must be between 1 and 100")
        continue

    try:

        url = f"{BASE_URL}/{product_id}"

        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product Not Found")

            log_error(
                "lookup_product",
                "HTTPError",
                f"404 Not Found for product ID {product_id}"
            )

        else:

            product = response.json()

            print(
                product["title"],
                "- $",
                product["price"]
            )

    except requests.exceptions.ConnectionError:
        print("Connection failed.")
        log_error(
            "lookup_product",
            "ConnectionError",
            "Network failure"
        )

    except requests.exceptions.Timeout:
        print("Timeout occurred.")
        log_error(
            "lookup_product",
            "Timeout",
            "Slow response"
        )

    except Exception as e:
        print(e)
        log_error(
            "lookup_product",
            type(e).__name__,
            str(e)
        )

 
```
